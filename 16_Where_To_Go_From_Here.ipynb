```python

A = np.array([[1,2],[3,4]])
B = np.array([[5],[6]])

print("Matrix Multiplication")
print(A @ B)
```

# 9. Step 5: Join the AI Community

## Pentingnya Komunitas

Banyak perkembangan AI terjadi melalui diskusi komunitas.

Manfaat:

* Belajar tren terbaru
* Berdiskusi dengan praktisi
* Menemukan peluang proyek
* Menemukan peluang kerja


```python

community_growth = np.cumsum(np.random.randint(1,8,20))

plt.figure(figsize=(8,4))
plt.plot(community_growth)
plt.title("Networking Growth Illustration")
plt.xlabel("Month")
plt.ylabel("Connections")
plt.show()
```

# 12. Step 7: Acquire Access to a GPU

## Mengapa GPU?

Training model besar membutuhkan komputasi yang tinggi.

GPU dapat melakukan operasi matriks secara paralel sehingga jauh lebih cepat dibanding CPU.


```python

hardware = ["CPU","GPU"]

speed = [1,20]

plt.figure(figsize=(6,4))
plt.bar(hardware,speed)
plt.title("Relative Training Speed")
plt.ylabel("Relative Speed")
plt.show()
```
