Pada chapter ini neural network mulai menerima banyak input sekaligus. Konsep ini lebih realistis karena sebagian besar masalah machine learning memiliki banyak fitur yang mempengaruhi hasil prediksi.


```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

print("Chapter 2 Loaded")
```


# 1. Mengapa Banyak Input?

Pada dunia nyata, prediksi jarang hanya bergantung pada satu variabel.

Contoh prediksi kebugaran:

* Jumlah langkah kaki
* Jam tidur
* Konsumsi kalori

Ketiga faktor tersebut dapat digunakan secara bersamaan untuk menghasilkan prediksi yang lebih baik.


```python

inputs = np.array([8.5, 7.0, 3.0])

print(inputs)
```


# 2. Konsep Multiple Inputs

Setiap input memiliki weight masing masing.

Misalnya:

* Langkah kaki → weight 0.1
* Jam tidur → weight 0.2
* Kalori → weight 0.05

Neural network akan menggabungkan seluruh kontribusi input tersebut.


```python

weights = np.array([0.1, 0.2, 0.05])

print(weights)
```


# 3. Weighted Sum

Prediksi dihitung menggunakan penjumlahan hasil perkalian antara input dan weight.

Formula:

prediction = Σ(input × weight)

Konsep ini menjadi dasar hampir seluruh neural network modern.


```python

inputs = np.array([8.5, 7.0, 3.0])
weights = np.array([0.1, 0.2, 0.05])

prediction = (inputs * weights).sum()

print("Prediction:", prediction)
```


# 4. Menggunakan Dot Product

NumPy menyediakan fungsi dot product yang membuat perhitungan lebih sederhana.

Dot product adalah operasi yang sangat penting dalam deep learning.


```python

prediction = np.dot(inputs, weights)

print("Prediction:", prediction)
```


# 5. Visualisasi Kontribusi Setiap Input

Setiap input memberikan kontribusi berbeda terhadap prediksi akhir.


```python

contribution = inputs * weights

plt.figure(figsize=(7,4))
plt.bar(["Steps","Sleep","Calories"], contribution)
plt.title("Contribution of Each Input")
plt.ylabel("Contribution")
plt.show()
```


# 6. Memahami Vector

Dalam machine learning, kumpulan fitur biasanya direpresentasikan sebagai vector.

Contoh:

[8.5, 7.0, 3.0]

adalah sebuah vector dengan tiga elemen.


```python

vector = np.array([8.5,7.0,3.0])

print("Dimension:", vector.shape)
```


# 7. Hubungan Vector dan Neural Network

Input layer pada neural network sebenarnya merupakan kumpulan nilai dalam bentuk vector.

Semakin banyak fitur yang digunakan, semakin panjang vector input yang dimiliki.



# 8. Studi Kasus Prediksi Sederhana

Misalkan:

* Langkah kaki
* Jam tidur
* Konsumsi air

digunakan untuk memprediksi tingkat kesehatan.


```python

health_features = np.array([10,8,6])

weights = np.array([0.3,0.4,0.2])

prediction = np.dot(health_features, weights)

print(prediction)
```


# 9. Eksperimen Mengubah Weight

Perubahan weight akan mengubah hasil prediksi.


```python

inputs = np.array([8.5,7.0,3.0])

for w in [0.1,0.2,0.3,0.4]:
    weights = np.array([w,w,w])
    prediction = np.dot(inputs, weights)
    print(f"Weight={w}, Prediction={prediction:.3f}")
```


# 10. Visualisasi Pengaruh Weight

Semakin besar weight, semakin besar hasil prediksi.


```python

weight_values = np.linspace(0,1,50)

predictions = []

for w in weight_values:
    predictions.append(np.dot(inputs, [w,w,w]))

plt.figure(figsize=(8,4))
plt.plot(weight_values, predictions)
plt.title("Weight vs Prediction")
plt.xlabel("Weight")
plt.ylabel("Prediction")
plt.grid(True)
plt.show()
```


# 11. Multiple Outputs

Neural network juga dapat menghasilkan lebih dari satu output.

Contoh:

Input:
* Data kesehatan

Output:
* Tingkat kebugaran
* Risiko penyakit
* Kualitas hidup

Konsep ini akan dibahas lebih lanjut pada chapter berikutnya.


```python

inputs = np.array([8.5,7,3])

weights_matrix = np.array([
    [0.1,0.2,0.0],
    [0.1,0.1,0.1],
    [0.3,0.1,0.2]
])

outputs = np.dot(weights_matrix, inputs)

print(outputs)
```


# 12. Mengapa Dot Product Penting?

Hampir seluruh operasi forward propagation pada neural network modern menggunakan dot product.

Keunggulan:

* Cepat
* Efisien
* Mudah dioptimalkan
* Cocok untuk GPU
