```python

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

print("Chapter 3 Loaded")
```


# 1. Mengapa Multiple Outputs?

Banyak masalah memiliki lebih dari satu target.

Contoh sistem kesehatan:

Input:
* Langkah kaki
* Jam tidur
* Asupan kalori

Output:
* Tingkat kebugaran
* Risiko penyakit
* Kualitas kesehatan

Karena itu neural network harus mampu menghasilkan beberapa output sekaligus.


```python

inputs = np.array([8.5, 7.0, 3.0])

print(inputs)
```


# 2. Weight Matrix

Untuk menghasilkan banyak output, kita tidak lagi menggunakan satu vector weight.

Kita menggunakan matrix weight.

Setiap baris mewakili satu output.


```python

weights = np.array([
    [0.1, 0.1, -0.3],
    [0.1, 0.2, 0.0],
    [0.0, 1.3, 0.1]
])

print(weights)
```


# 3. Forward Propagation dengan Multiple Outputs

Prediksi diperoleh menggunakan matrix multiplication.

Formula:

outputs = weights × inputs


```python

inputs = np.array([8.5,7.0,3.0])

weights = np.array([
    [0.1,0.1,-0.3],
    [0.1,0.2,0.0],
    [0.0,1.3,0.1]
])

prediction = np.dot(weights, inputs)

print(prediction)
```


# 4. Interpretasi Output

Misalkan hasil prediksi:

Output 1 = Fitness Score

Output 2 = Performance Score

Output 3 = Injury Risk Score

Setiap output dihitung dari kombinasi input yang sama tetapi menggunakan weight berbeda.



# 5. Visualisasi Output

Kita dapat melihat bagaimana setiap output memiliki nilai yang berbeda.


```python

labels = ["Output 1", "Output 2", "Output 3"]

plt.figure(figsize=(7,4))
plt.bar(labels, prediction)
plt.title("Multiple Outputs")
plt.show()
```


# 6. Matrix Multiplication

Matrix multiplication merupakan fondasi deep learning.

Hampir semua operasi pada neural network modern dapat ditulis sebagai perkalian matriks.


```python

A = np.array([
    [1,2,3],
    [4,5,6]
])

B = np.array([1,2,3])

print(np.dot(A,B))
```


# 7. Hubungan Matrix dan Neural Network

Setiap neuron output memiliki sekumpulan weight sendiri.

Karena itu weight disimpan dalam bentuk matrix.

Semakin banyak output, semakin besar ukuran matrix.


```python

inputs = 3
outputs = 5

weight_matrix = np.random.randn(outputs, inputs)

print(weight_matrix.shape)
```


# 8. Studi Kasus Prediksi Atlet

Input:

* Latihan
* Tidur
* Nutrisi

Output:

* Kecepatan
* Daya tahan
* Risiko cedera


```python

athlete = np.array([9,8,7])

weights = np.array([
    [0.5,0.2,0.1],
    [0.1,0.6,0.2],
    [-0.2,-0.1,0.5]
])

scores = np.dot(weights, athlete)

print(scores)
```


# 9. Eksperimen Mengubah Input

Perubahan input akan mempengaruhi seluruh output.


```python

training_values = np.arange(0,11)

output_scores = []

for value in training_values:
    sample = np.array([value,8,7])
    score = np.dot(weights, sample)[0]
    output_scores.append(score)

plt.figure(figsize=(8,4))
plt.plot(training_values, output_scores)
plt.title("Training Level vs Output 1")
plt.xlabel("Training")
plt.ylabel("Output Score")
plt.grid(True)
plt.show()
```


# 10. Multiple Inputs dan Multiple Outputs

Pada titik ini neural network telah mampu:

* Banyak input
* Banyak output

Ini merupakan bentuk dasar jaringan saraf yang lebih realistis.


```python

inputs = np.array([8.5,7,3])

weights = np.random.randn(4,3)

outputs = np.dot(weights, inputs)

print(outputs)
```


# 11. Kelebihan Multiple Outputs
