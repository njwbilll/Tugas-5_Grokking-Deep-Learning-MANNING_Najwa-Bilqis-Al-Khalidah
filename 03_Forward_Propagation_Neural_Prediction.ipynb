# Chapter 3: Forward Propagation: Pengantar Prediksi Jaringan Saraf Tiruan

**Referensi:** Grokking Deep Learning, Andrew W. Trask (Manning Publications, 2019)

## Ringkasan Bab

Bab ketiga ini membahas mekanika internal paling mendasar dari jaringan saraf tiruan saat melakukan estimasi, yang dikenal dengan istilah propagasi maju atau forward propagation. Jaringan saraf tiruan pada intinya adalah sebuah alat hitung yang menerima informasi dari dunia luar, mengalikan informasi tersebut dengan parameter internal yang disebut bobot, lalu menghasilkan keluaran berupa prediksi. Proses ini meniru bagaimana sistem saraf makhluk hidup mengalirkan sinyal dari reseptor menuju efektor secara searah.

## Teori Mendalam Propagasi Maju

### A. Anatomi Hubungan Input dan Bobot
Operasi dasar dari sebuah neuron tunggal adalah kombinasi linear. Bobot merepresentasikan kekuatan koneksi antar lapiasan jaringan. Jika bobot bernilai tinggi, maka variabel input tersebut memiliki pengaruh dominan terhadap keputusan akhir model. Sebaliknya, jika bobot mendekati nol, informasi input tersebut diabaikan oleh jaringan saraf.

### B. Kompleksitas Multi Input dan Multi Output
Dunia nyata memiliki banyak variabel interdependen. Model tingkat lanjut memperluas struktur data menjadi arsitektur multi input, di mana beberapa parameter diakumulasikan menggunakan operasi perkalian titik atau dot product. Struktur ini dapat dikembangkan lebih lanjut menjadi sistem multi output untuk menghasilkan spektrum prediksi yang kaya dari satu sumber informasi tunggal.

### C. Konsep Lapisan Tersembunyi atau Hidden Layers
Kekuatan sejati dari deep learning terletak pada penumpukan lapisan lapisan perantara di antara input dan output asli. Lapisan tersembunyi ini bertindak sebagai pembuat profil otomatis yang mengekstraksi representasi abstrak dari data mentah, memungkinkan jaringan saraf memahami pola nonlinear yang rumit.

### Tabel Koding 1: Persiapan Pustaka Komputasi Vektor
Sebelum membangun arsitektur, pustaka NumPy dimuat sebagai instrumen utama untuk mengelola aljabar linear berkinerja tinggi.

```python
import numpy as np
print("Pustaka NumPy telah aktif untuk mendukung operasi matematika maju")
```

### Tabel Koding 2: Jaringan Saraf Paling Sederhana Tunggal
Implementasi awal ini menunjukkan struktur fundamental di mana satu nilai input dipetakan langsung ke satu output melalui perkalian bobot tunggal.

```python
def jaringan_saraf_tunggal(data_input, bobot_reguler):
    prediksi_akhir = data_input * bobot_reguler
    return prediksi_akhir

jumlah_indeks_input = 2.5
nilai_bobot_awal = 0.3
hasil_prediksi = jaringan_saraf_tunggal(jumlah_indeks_input, nilai_bobot_awal)
print("Hasil estimasi tunggal jaringan:", hasil_prediksi)
```

### Tabel Koding 3: Mekanisme Multi Input Menggunakan Struktur Perulangan
Ketika dihadapkan pada beberapa variabel input sekaligus, jaringan harus mengalikan setiap elemen secara berpasangan lalu menjumlahkannya.

```python
def kalkulasi_multi_input_manual(vektor_input, vektor_bobot):
    total_prediksi = 0
    for i in range(len(vektor_input)):
        total_prediksi += (vektor_input[i] * vektor_bobot[i])
    return total_prediksi

informasi_variabel = [8.5, 0.65, 1.2]
konfigurasi_bobot = [0.1, 0.2, 1.3]
estimasi_manual = kalkulasi_multi_input_manual(informasi_variabel, konfigurasi_bobot)
print("Hasil kombinasi linear manual:", estimasi_manual)
```

### Tabel Koding 4: Optimalisasi Multi Input dengan Dot Product NumPy
Untuk efisiensi komputasi, perulangan di atas digantikan oleh fungsi dot product bawaan NumPy yang mengeksekusi operasi secara paralel.

```python
array_input = np.array([8.5, 0.65, 1.2])
array_bobot = np.array([0.1, 0.2, 1.3])
estimasi_vektorisasi = np.dot(array_input, array_bobot)
print("Hasil kombinasi linear via dot product:", estimasi_vektorisasi)
```

### Tabel Koding 5: Jaringan Saraf dengan Struktur Multi Output
Dalam skenario ini, satu sumber data tunggal digunakan untuk memprediksi beberapa fenomena terpisah sekaligus secara elementwise.

```python
def kalkulasi_multi_output(data_tunggal, matriks_bobot):
    kumpulan_estimasi = data_tunggal * matriks_bobot
    return kumpulan_estimasi

input_spesifik = 0.75
bobot_beragam = np.array([0.3, 0.6, 0.9])
daftar_prediksi = kalkulasi_multi_output(input_spesifik, bobot_beragam)
print("Kumpulan matriks output hasil prediksi:", daftar_prediksi)
```

### Tabel Koding 6: Arsitektur Gabungan Multi Input dan Multi Output
Mengkombinasikan kedua konsep sebelumnya menghasilkan matriks bobot dua dimensi yang menghubungkan sekelompok input dengan sekelompok output.

```python
matriks_koneksi_bobot = np.array([
    [0.1, 0.1, 0.3],
    [0.1, 0.2, 0.8],
    [0.4, 0.4, 0.1]
])

vektor_fitur_input = np.array([10.3, 1.5, 0.6])
output_kombinasi = np.dot(matriks_koneksi_bobot, vektor_fitur_input)
print("Hasil konvolusi linear multi input multi output:", output_kombinasi)
```

### Tabel Koding 7: Pembentukan Struktur Lapisan Tersembunyi Berjenjang
Di sini kita mulai menumpuk jaringan saraf dengan menambahkan satu lapisan antara untuk menangkap esensi data secara berjenjang.

```python
bobot_input_ke_tersembunyi = np.array([
    [0.1, 0.2, 0.4],
    [0.4, 0.1, 0.2],
    [0.2, 0.3, 0.1]
])

bobot_tersembunyi_ke_output = np.array([
    [0.3, 0.1, 0.1],
    [0.1, 0.4, 0.2],
    [0.1, 0.1, 0.5]
])

print("Arsitektur dua lapis berbobot positif berhasil disiapkan")
```

### Tabel Koding 8: Eksekusi Manual Aliran Forward Propagation Multi Layer
Melakukan kalkulasi bertahap dari lapisan input menuju hidden layer, lalu melanjutkannya dari hidden layer menuju lapisan output final.

```python
fitur_mentah = np.array([2.0, 1.0, 0.5])

aktivasi_tersembunyi = np.dot(bobot_input_ke_tersembunyi, fitur_mentah)
print("Nilai internal pada lapisan tersembunyi:", aktivasi_tersembunyi)

prediksi_ujung_akhir = np.dot(bobot_tersembunyi_ke_output, aktivasi_tersembunyi)
print("Nilai konklusi pada lapisan output akhir:", prediksi_ujung_akhir)
```

### Tabel Koding 9: Enkapsulasi Fungsi Jaringan Saraf Jauh
Seluruh rangkaian logika berjenjang tersebut digabungkan ke dalam sebuah fungsi modular yang bersih agar mudah dipanggil berulang kali.

```python
def prediksi_jaringan_mendalam(input_data, matriks_awal, matriks_kedua):
    lapisan_satu = np.dot(matriks_awal, input_data)
    lapisan_dua = np.dot(matriks_kedua, lapisan_satu)
    return lapisan_dua

hasil_modular = prediksi_jaringan_mendalam(fitur_mentah, bobot_input_ke_tersembunyi, bobot_tersembunyi_ke_output)
print("Output dari fungsi enkapsulasi mendalam:", hasil_modular)
```

### Tabel Koding 10: Pemrosesan Sekaligus Banyak Sampel Data
Sebagai tingkat lanjut, kita menguji fungsi untuk menerima matriks data yang berisi beberapa baris sampel guna menghasilkan sekumpulan prediksi sekaligus secara simultan.

```python
kumpulan_sampel_input = np.array([
    [2.0, 1.0, 0.5],
    [1.0, 3.0, 0.2],
    [0.5, 0.5, 4.0]
])

for indeks_sampel, sampel in enumerate(kumpulan_sampel_input):
    output_sampel = prediksi_jaringan_mendalam(sampel, bobot_input_ke_tersembunyi, bobot_tersembunyi_ke_output)
    print(f"Hasil prediksi untuk sampel ke {indeks_sampel + 1}: {output_sampel}")

print("\nSeluruh proses pengujian komputasi forward propagation bab tiga selesai dieksekusi")
```
