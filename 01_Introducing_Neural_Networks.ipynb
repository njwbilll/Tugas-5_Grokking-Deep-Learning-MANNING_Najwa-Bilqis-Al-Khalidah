```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

print("Environment Ready")
```


# 1. Apa Itu Neural Network?

Neural Network adalah model komputasi yang terinspirasi dari cara kerja neuron pada otak manusia.

Tujuan utamanya adalah mempelajari hubungan antara input dan output.

Contoh:

Input:
* Jumlah langkah kaki

Output:
* Prediksi tingkat kebugaran

Input:
* Nilai ujian
* Jam belajar

Output:
* Prediksi kelulusan

Pada dasarnya neural network mencoba menemukan pola dari data.



# 2. Intuisi Prediksi

Bayangkan kita ingin memprediksi tingkat kebugaran seseorang berdasarkan jumlah langkah kaki per hari.

Jika seseorang berjalan lebih banyak, kemungkinan tingkat kebugarannya juga lebih baik.

Neural network mencoba mempelajari hubungan tersebut menggunakan angka yang disebut weight.


```python

steps = 8.5
weight = 0.1

prediction = steps * weight

print("Prediction:", prediction)
```


# 3. Konsep Weight

Weight adalah parameter yang menunjukkan seberapa besar pengaruh suatu input terhadap output.

Jika weight besar:

Input memiliki pengaruh kuat.

Jika weight kecil:

Input memiliki pengaruh lemah.

Seluruh proses pembelajaran neural network pada dasarnya adalah proses mencari weight terbaik.


```python

weights = [0.05, 0.1, 0.2, 0.5]

steps = 8.5

for w in weights:
    print(f"Weight={w}, Prediction={steps*w}")
```


# 4. Forward Propagation

Forward Propagation adalah proses menghitung output dari input yang diberikan.

Formula sederhana:

prediction = input × weight

Inilah operasi paling dasar dalam neural network.


```python

input_value = 10
weight = 0.8

prediction = input_value * weight

print(prediction)
```


# 5. Visualisasi Hubungan Input dan Prediksi

Ketika weight tetap, peningkatan input akan meningkatkan nilai prediksi.


```python

inputs = np.arange(0,20)

weight = 0.5

predictions = inputs * weight

plt.figure(figsize=(8,4))
plt.plot(inputs,predictions)
plt.title("Input vs Prediction")
plt.xlabel("Input")
plt.ylabel("Prediction")
plt.grid(True)
plt.show()
```


# 6. Multiple Inputs

Pada dunia nyata biasanya terdapat lebih dari satu input.

Contoh:

* Jumlah langkah kaki
* Jam tidur
* Kalori yang dikonsumsi

Masing masing input memiliki weight sendiri.


```python

inputs = np.array([8.5, 7.0, 3.0])
weights = np.array([0.1, 0.2, 0.05])

prediction = np.dot(inputs, weights)

print("Prediction:", prediction)
```


# 7. Dot Product

Dot Product merupakan operasi matematika yang sangat penting dalam deep learning.

Formula:

(a1 × b1) + (a2 × b2) + ...

Hampir seluruh neural network modern menggunakan operasi ini.


```python

a = np.array([1,2,3])
b = np.array([4,5,6])

print(np.dot(a,b))
```


# 8. Mengapa Neural Network Bisa Belajar?

Neural network dapat belajar karena weight dapat diubah.

Jika prediksi salah:

Weight diperbaiki.

Jika prediksi benar:

Weight dipertahankan.

Proses perbaikan weight inilah yang nantinya menjadi inti dari training.



# 9. Studi Kasus Sederhana

Misalkan:

Input = jumlah langkah kaki

Target = tingkat kebugaran aktual

Kita ingin mencari weight yang menghasilkan prediksi paling mendekati target.


```python

steps = 8.5
target = 1.0

weights = np.linspace(0,1,20)

predictions = steps * weights

errors = (predictions - target) ** 2

best_weight = weights[np.argmin(errors)]

print("Best Weight:", round(best_weight,4))
```


# 10. Visualisasi Error

Semakin kecil error, semakin baik model.


```python

plt.figure(figsize=(8,4))
plt.plot(weights, errors)
plt.title("Weight vs Error")
plt.xlabel("Weight")
plt.ylabel("Squared Error")
plt.grid(True)
plt.show()
```


# 11. Interpretasi Error

Error menunjukkan seberapa jauh prediksi dari nilai sebenarnya.

Jika:

Prediction = Target

Maka error = 0

Tujuan training adalah meminimalkan error.
